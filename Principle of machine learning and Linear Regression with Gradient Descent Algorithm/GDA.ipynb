```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]

y=[2,4.1,6.3,10.2, 10.5, 12.1,14.1]

w = [1,1]

plt.plot(x,y, 'ro')

plt.xlabel('X')

plt.ylabel('Y')

x0= np.linspace(start=1, stop=10, num=50)

y0= w[0]+w[1]*x0

plt.plot(x0,y0)

plt.show()
```

```python
# hàm y = f(x) # w: tham số
def f(x, w):
    return w[0] + w[1] * x

# tính hàm loss
def loss(x, y, w):
    d = 0
    for i in range(len(x)):
        d += (y[i] - f(x[i], w)) ** 2
    return d / (2 * len(x))

# tính đạo hàm tại điểm w[0] và w[1]
def derivative(x, y, w):
    d0 = 0
    d1 = 0
    for i in range(len(x)):
        d0 += f(x[i], w) - y[i]
        d1 += (f(x[i], w) - y[i]) * x[i]
    return d0 / len(x), d1 / len(x)
```

```python


# training
epoch = 20
learning_rate = 0.01
w = [1, 1]  # y = x + 1
los_old = 0

for i in range(epoch):
    # hiển thị đồ thị
    plt.plot(x, y, 'ro')
    plt.xlabel('X')
    plt.ylabel('Y')

    x0 = np.linspace(start=1, stop=10, num=50)
    y0 = f(x0, w)  # sử dụng hàm f để tính y0
    plt.plot(x0, y0)
    plt.show()

    # kiểm tra xem loss có giảm không
    los = loss(x, y, w)
    print('epoch', i + 1, ':')
    print('Loss:', los, 'Previous Loss:', los_old)
    print('Weights:', w)

    if abs(los - los_old) < 0.0001:
        break

    los_old = los

    # cập nhật
    a, b = derivative(x, y, w)  # tính đạo hàm
    w[0] = w[0] - a * learning_rate
    w[1] = w[1] - b * learning_rate
```
