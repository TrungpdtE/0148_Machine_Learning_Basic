```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([2, 4.1, 6.3, 10.2, 10.5, 12.1, 14.1])

# Tham số khởi tạo
w = [1, 1]

# Hiển thị dữ liệu
plt.plot(x, y, 'ro', label='Dữ liệu')
plt.xlabel('X')
plt.ylabel('Y')

# Vẽ đường hồi quy ban đầu
x0 = np.linspace(start=1, stop=10, num=50)
y0 = w[0] + w[1] * x0
plt.plot(x0, y0, label='Mô hình dự đoán')
plt.legend()
plt.show()
```

```python
# Hàm y = f(x)
def f(x, w):
    return w[0] + w[1] * x

# Tính hàm loss
def loss(x, y, w):
    d = 0
    for i in range(len(x)):
        d += (y[i] - f(x[i], w))**2
    return d / (2 * len(x))
```

```python
# Tính đạo hàm
def derivative(x, y, w):
    d0 = 0  # Đạo hàm theo w[0]
    d1 = 0  # Đạo hàm theo w[1]
    for i in range(len(x)):
        d0 += f(x[i], w) - y[i]
        d1 += (f(x[i], w) - y[i]) * x[i]
    return d0 / len(x), d1 / len(x)
```

```python
# Huấn luyện
epoch = 20
learning_rate = 0.01
w = [1, 1]  # Trọng số khởi tạo
los_old = 0

for i in range(epoch):
    # Hiển thị đồ thị
    plt.plot(x, y, 'ro', label='Dữ liệu')
    plt.xlabel('X')
    plt.ylabel('Y')

    x0 = np.linspace(start=1, stop=10, num=50)
    y0 = f(x0, w)
    plt.plot(x0, y0, label='Mô hình dự đoán')
    plt.legend()
    plt.pause(0.5)  # Tạm dừng để xem đồ thị

    # Kiểm tra xem loss có giảm không
    los = loss(x, y, w)
    print('Epoch', i + 1, ':')
    print('Loss:', los, 'Old Loss:', los_old)
    print('Weights:', w)

    if abs(los - los_old) < 0.0001:
        break

    los_old = los

    # Cập nhật weights
    d0, d1 = derivative(x, y, w)  # Tính đạo hàm
    w[0] = w[0] - d0 * learning_rate
    w[1] = w[1] - d1 * learning_rate

plt.show()
```
